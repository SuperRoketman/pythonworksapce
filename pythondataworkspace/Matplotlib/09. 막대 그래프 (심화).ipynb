# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # /////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ...
```

더 많은 hatch style 은 matplotlib 공식 홈페이지에서 더 찾아볼 수 있음.

```python
bar = plt.bar(labels, values)
plt.ylim(180, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='r')
```
